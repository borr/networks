# Statistical Characterization of Network Traffic

Interdomain traffic patterns significantly impact the evolution of interdomain topology and economics, Internet pricing, and policy considerations (e.g., network neutrality), thus we need more knowledge of the IP Networks' Interdomain Traffic Matrix (ITM) and of its dynamics. In this notebook, we take first steps towards inferring some statistical properties of the interdomain traffic matrix and rely on passive flow data collected at router-interface levels. 

In this preliminary work, the focus is on spatial properties of the ITM, in particular, this is a characterization of the visible portion of the customer-customer(@RTR-ASN) traffic matrix.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats
%pylab inline
```

Heavy tailed distributions are commonly observed in IP Networks, it is likely that we obtain such distributions in our dataset as well. Create a flow dataset by combining aggregate flows in the ingress and egress directions:

```python
master = pd.read_pickle("master")
ingress = master[master['direction'] == 1]
egress = master[master['direction'] == 2]

# define measures to characterize
measures = ['bytes','speed','hopspeed','hops']

inflows = ingress.groupby(['inASN','inRTR','egASN','egRTR'])[measures].mean()
egflows = egress.groupby(['inASN','inRTR','egASN','egRTR'])[measures].mean()
flows = inflows.add(egflows).div(2)
idx = flows.isnull()
flows[idx] = egflows[idx]
idx = flows.isnull()
flows[idx] = inflows[idx]
flows = flows.reset_index()
```

```python
flows.to_pickle("flows")
```

Evidence for heavy-tailed distributions is observed in the majority of the rows – the top 5% of the destination RTR-ASNs account for over 95% of the traffic. Refer to the tail as the traffic sent to the top 10% of destinations by the corresponding customer. Also, only consider sources sending atleast 500 flows and from these ASNs extract top ASNs/Routers by aggregate flow count.

```python
cust_count = flows.groupby(['inASN','inRTR'])['speed'].count()
cust_count.sort(ascending=False)
```

Keep those customers with a minimum 500 aggregated-flows which comes out to be about be the top 150:

```python
cust_list = cust_count[:150]
plt.hist(cust_count)
```

## Case 1: Keep Noise

Here are the traffic distributions (CCDF) of three heavy hitters, we may obtain some distribution parameters by fitting the respective distributions ('Lognormal','Pareto') and then assess goodness of fit with the Kolmogorov-Smirnov test, although we are interested in characterizing the tail of these distributions, ignoring the values in the main body. This is due to the noisy nature of netflow sampling consisting of small values in the body.

```python
data = pd.DataFrame(cust_list[:3],columns=['count_flows']).reset_index().merge(flows)
groups = data.groupby(['inASN','inRTR']).groups
```

```python
dist = {'lognormal':{'params':{},'ks':{}},'pareto':{'params':{},'ks':{}}}

for key,value in groups.iteritems():
    speed = data.loc[value]['speed']
    params = stats.lognorm.fit(speed,loc=0)
    dist['lognormal']['params'][key] = params
    params = stats.pareto.fit(speed)
    dist['pareto']['params'][key] = params
```

```python
for key,value in groups.iteritems():
    speed = data.loc[value]['speed']
    dist['lognormal']['ks'][key] = stats.kstest(speed,'lognorm',args=dist['lognormal']['params'][key]) 
    dist['pareto']['ks'][key] = stats.kstest(speed,'pareto',args=dist['pareto']['params'][key])
```

The tail of the distributions in the Figure below can be modeled as Pareto (CCDF in log-log scale resembles a straight line) OR this distribution can be modeled as LogNormal. The tail of LogNormal decays faster than the tail of Pareto, so there is a higher probability of observing heavy hitters that follow a Pareto distribution rather than a LogNormal. Here are the actual empirical distributions along with the fitted LogNormal/Pareto equivalents:

```python
fig = plt.figure(figsize = (16,8))
ax1 = fig.add_subplot(111)
ax1.set_xscale('log')
ax1.set_yscale('log')

X = {'lognormal':{}, 'pareto':{}}
lognorm = dist['lognormal']['params']
pareto = dist['pareto']['params']

for key,value in groups.iteritems():
    speed = data.loc[value]['speed']
    ecdf = sm.distributions.ECDF(speed)
    X['lognormal'][key] = stats.lognorm.rvs(lognorm[key][0], loc=lognorm[key][1], scale=lognorm[key][2], size=1000)
    X['pareto'][key] = stats.pareto.rvs(pareto[key][0], loc=pareto[key][1], scale=pareto[key][2], size=1000)
    ecdf_lgn = sm.distributions.ECDF(X['lognormal'][key])
    ecdf_par = sm.distributions.ECDF(X['pareto'][key])
    x = np.linspace(min(speed), max(speed))
    y = ecdf(x)
    y1 = ecdf_lgn(x)
    y2 = ecdf_par(x)
    ax1.plot(x,1-y,label = key,linewidth=2)
    ax1.plot(x,1-y1,label = str(key)+" lognorm",marker='+')
    ax1.plot(x,1-y2,label = str(key)+" pareto",marker='*')
    legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
```

## Case 2: Remove Noise

Due to the noisy nature of netflow sampling comprising smaller values in the body, we filter the dataset retaining the top source customers (reference); then filter the top 10 percentile destinations as representative of the heavy tails. Define a function for getting the top percentile values:

```python
per = 0.9
```

```python
def f(df,per=per,measure='speed'):
    return df[df[measure]>df[measure].quantile(per)]
```

```python
data_top = pd.DataFrame(cust_list[:10],columns=['count_flows']).reset_index().merge(flows).groupby(["inASN","inRTR"], group_keys=False).apply(f)
groups_top = data_top.groupby(['inASN','inRTR']).groups
```

Here are (truncated) distributions with heavy tails retained for the top 10 heavy hitters:

```python
fig = plt.figure(figsize = (16,8))
ax1 = fig.add_subplot(111)
ax1.set_xscale('log')
ax1.set_yscale('log')

for key,value in groups_top.iteritems():
    speed = data_top.loc[value]['speed']
    ecdf = sm.distributions.ECDF(speed)
    x = np.linspace(min(speed), max(speed))
    y = ecdf(x)
    ax1.plot(x,1-y,label = key,linewidth=2)
    legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
```

### a. No Sampling

Although, these are truncated versions of the distributions, assume that these is *complete* and do an estimation of the parameters. To generate synthetic distributions of sourced traffic, we need to know the nature of the distribution and the associated parameters, in particular, the “shape” parameter of these two distributions in the heavy tails. Here are those distribution parameters obtained by fitting the respective distributions ('Lognormal','Pareto'):

Assess goodness of fit with the Kolmogorov-Smirnov Test:

```python
dist = {'lognormal':{'params':{},'ks':{}},'pareto':{'params':{},'ks':{}}}

for key,value in groups_top.iteritems():
    speed = data_top.loc[value]['speed']
    params = stats.lognorm.fit(speed,loc=0)
    dist['lognormal']['params'][key] = params
    params = stats.pareto.fit(speed)
    dist['pareto']['params'][key] = params
```

```python
for key,value in groups_top.iteritems():
    speed = data_top.loc[value]['speed']
    dist['lognormal']['ks'][key] = stats.kstest(speed,'lognorm',args=dist['lognormal']['params'][key]) 
    dist['pareto']['ks'][key] = stats.kstest(speed,'pareto',args=dist['pareto']['params'][key])
```

### b. Sampling

With the distribution parameters obtained by fitting the respective distributions ('Lognormal','Pareto'), sample some data from the two candidate distributions and threshold with the tail cutoffs on the list of all ASNs/RTRs:

```python
data = pd.DataFrame(cust_list,columns=['count_flows']).reset_index().merge(flows).groupby(["inASN","inRTR"], group_keys=False).apply(f)
cutoff = data.groupby(['inASN','inRTR']).quantile(per)[measures]
groups = data.groupby(['inASN','inRTR']).groups
```

```python
dist = {'lognormal':{'params':{},'ks':{}},'pareto':{'params':{},'ks':{}}}

for key,value in groups.iteritems():
    speed = data.loc[value]['speed']
    params = stats.lognorm.fit(speed,loc=0)
    dist['lognormal']['params'][key] = params
    params = stats.pareto.fit(speed)
    dist['pareto']['params'][key] = params
```

Let X be the examined sample and F be the empirical distribution of X. The tail of X consists of all values in the top 10 percentile of the distribution, i.e., the values above some “tail threshold” $τ$. Let F be a candidate distribution (Log-Normal or Pareto) that we try to fit to the tail of F. From the candidate distribution F we generate a sample X' and finally generate a sample like :

$ \hat{X} = $ {$ X′: X′ < τ , X : X ≥ τ$ }

Then finally, perform a two-sample Kolmogorov-Smirnov test to check whether the distributions differ in the tails:

```python
sample = {'lognormal':{}, 'pareto':{}}
X = {'lognormal':{}, 'pareto':{}}

for key,value in dist['lognormal']['params'].iteritems():
    X['lognormal'][key] = stats.lognorm.rvs(value[0], loc=value[1], scale=value[2], size=250)
    sample['lognormal'][key] = X['lognormal'][key][X['lognormal'][key]<=cutoff.ix[key]['speed']]
    speed = data.loc[groups[key]]['speed']
    sample['lognormal'][key] = pd.Series(np.append(sample['lognormal'][key], speed))

for key,value in dist['pareto']['params'].iteritems():
    X['pareto'][key] = stats.pareto.rvs(value[0], loc=value[1], scale=value[2], size=200)
    sample['pareto'][key] = X['pareto'][key][X['pareto'][key]<=cutoff.ix[key]['speed']]
    speed = data.loc[groups[key]]['speed']
    sample['pareto'][key] = pd.Series(np.append(sample['pareto'][key],speed))
```

```python
dist['pareto']['fit'] = {}
dist['lognormal']['fit'] = {}

for key in cust_list.index:
    dist['pareto']['fit'][key] = stats.ks_2samp(sample['pareto'][key], X['pareto'][key])
    dist['lognormal']['fit'][key] = stats.ks_2samp(sample['lognormal'][key], X['lognormal'][key])
```

Here is a characterization on all the customers for which distribution fits best. This is observed as *rejection* of the null hypothesis; implying p-values are greater than the specified cutoffs, threshold at $[0.8, 0.9]$ :

```python
thres = 0.9
pareto_list = []
lognormal_list = []

for key in cust_list.index:
    if dist['pareto']['fit'][key][1] > dist['lognormal']['fit'][key][1]:
        if dist['pareto']['fit'][key][1] > thres:
            pareto_list.append(key)
        else:
            pass
    else:
        if dist['lognormal']['fit'][key][1] > thres:
            lognormal_list.append(key)
```

### Distribution Parameters

Investigate whether the shape of the measured distributions depends on the Customer's traffic throughput, i.e., on the total traffic generated by that Customer. The shape of the Pareto distribution is represented by the $α$ parameter; lower values of $α$ indicate a heavier tail. For LogNormal, characterize the shape of the distribution using the coefficient of variation ($CoV$); a higher $CoV$ indicates a heavier tail. Plot distribution parameters VS throughput for the customers as per their characterization above:

```python
df = pd.DataFrame(data.groupby(['inASN','inRTR'])['speed'].sum())
df = df.join(pd.DataFrame(cust_count,columns = ['count']))
```

```python
dist['lognormal']['cv'] = {}
dist['pareto']['alpha'] = {}

for key in cust_list.index:
    dist['lognormal']['cv'][key] = np.sqrt(np.exp(dist['lognormal']['params'][key][0]**2)-1)
    dist['pareto']['alpha'][key] = dist['pareto']['params'][key][0]
```

```python
df['pareto'] = 0
df['lognormal'] = 0
df.loc[df.index.isin(pareto_list),"pareto"] = 1
df.loc[df.index.isin(lognormal_list),"lognormal"] = 1
```

```python
df = df.join(pd.DataFrame.from_dict(dist['lognormal'],'columns')['cv']).join(pd.DataFrame.from_dict(dist['pareto'],'columns')['alpha'])
```

```python
sub = df[df['lognormal'] == 1]

fig = plt.figure(figsize = (10,7))
ax1 = fig.add_subplot(111)
ax1.set_xscale('log')

ax1.scatter(sub['speed'],sub['cv'],s=50)
```

### Shape of Tail

To investigate why the generated traffic distribution follows a LogNormal tail for some customers, and a Pareto for others, use the Kolmogorov-Smirnov metric that indicates whether the tail is LogNormal-like or Pareto-like. Then define a measure $"D"$ to indicate how close the tail of the empirical distribution resembles a LogNormal VS Pareto.

$D = KS(lognormal)$ $/$ $KS(pareto)+KS(lognormal)$

Thus $D=0$ implies lognormal fit and conversely $D=1$ implies a pareto fit

```python
df['D'] = 0

for key in cust_list.index:
    speed = data.loc[groups[key]]['speed']
    par_fit = stats.ks_2samp(speed, X['pareto'][key][X['pareto'][key]>cutoff.ix[key]['speed']])
    log_fit = stats.ks_2samp(speed, X['lognormal'][key][X['lognormal'][key]>cutoff.ix[key]['speed']])
    df.ix[key, 'D'] = log_fit[0]/(log_fit[0]+par_fit[0])
```

Here is a plot between the "D" statistic and throughput with each point representing a customer that has been successfully characterised by the above methodology. A lot of points have values incresingly near $D = 0.5$ which is either because the empirical CDF lies between a lognormal and a pareto, or there isnt enough data for significant characterization.

```python
sub = df[df['lognormal'] == 1].append(df[df['pareto']==1])
                                      
fig = plt.figure(figsize = (10,7))
ax1 = fig.add_subplot(111)
ax1.set_xscale('log')

ax1.scatter(sub['speed'],sub['D'],s=sub['count'],alpha=0.5)
```

## Concluding Remarks

We extract some relevant statistical properties of the Inter-Customer Traffic Matrix by characterising the distribution of traffic flows between customers at a RTR-ASN level. Such properties also show a relationship with the aggregate characteristics of the customer. From a total of 3000 customers we shortlisted the top 150 based on the count of flows (>500). Out of these, there were close to 40 customers that were successfully characterised, with ~25 classified as lognormal and ~15 as pareto tailed.

A plot of the relevant distribution metrics ($ \alpha $ in case of pareto and $CoV$ in case of lognormal disstribution) with aggregate throughput, we see that there exist some trends for each of the customers. While the pareto $ \alpha $ decreases with increase in throughput, the $CoV$ shows an increasing behavior with throughput. Since the tails become heavier and the more popular destinations receive even more traffic which cause changes in the scale parameters of the fitted distributions.

There is also an investigation as to why certain tail shapes resemble pareto/lognormal. It might have something to do with higher throughput leading to lognormal-like tail shapes. [2] also associates this character with the problem of network congestion that leads to a truncation in the tail of the distribution from pareto-like to a lognormal-like. [2] also indicates that congestion is a problem for customers with higher throughput. Although, no evidence as such is provided, as there are no direct ways of quantification or measurement, however some intuitive explanations have been presented.
