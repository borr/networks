Based on this we may want to consider a segmentation of the customers on the basis of whether they send/receive a large fraction (say >95%) of their volume to a single Router/Pop/Region/Country. Here is some code for this:

```python
top = data.join(top_frame['speed','sum']).reset_index()
top['segment'] = 0
top.loc[top[('country','value1')] > 0.95, "segment"] = 1
top.loc[top[('region','value1')] > 0.95, "segment"] = 2
top.loc[top[('rtr','value1')] > 0.95, "segment"] = 3
top['country'] = top[('country','label1')]
top['rtr'] = top[('rtr','label1')]
top['pop'] = top[('pop','label1')]
top['region'] = top[('region','label1')]
top['hops'] = top[('c(hops)','label1')]
top = top[['country','region','hops','rtr','pop','unit_cost','wtd_hops','level_0','level_1','level_2','level_3','level_4','segment']]
```

Lets look at each segment in some detail:

```python
fig = plt.figure(figsize = (16,12))
ax1 = fig.add_subplot(221)
ax1.hist(np.array(top[top['segment'] == 0]['unit_cost']))
ax2 = fig.add_subplot(222)
ax2.hist(np.array(top[top['segment'] == 1]['unit_cost']))
ax3 = fig.add_subplot(223)
ax3.hist(np.array(top[top['segment'] == 2]['unit_cost']))
ax4 = fig.add_subplot(224)
ax4.hist(np.array(top[top['segment'] == 3]['unit_cost']))
```
