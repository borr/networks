# Regression Analysis

The unit cost for a customer is defined as : Total Cost/Total Speed

Model the unit cost of any customer connected at a router-interface (index). This unit cost is the cost incurred for transferring unit bps of traffic (sending + receiving) in accordance with some cost models (unknown).

```python
%pylab inline
import pandas as pd
import numpy as np
```

Load the dataset data created earlier.

```python
sparse_frame = pd.read_pickle("sparse_frame")
top_frame = pd.read_pickle("top_frame")
aggregate_frame = pd.read_pickle("aggregate_frame")
```

```python
sparse_frame.sortlevel(level = 0,axis=1,inplace = True)
top_frame.sortlevel(level = 0,axis=1,inplace = True)
```

Create "unit_cost" -- Dependent Variable

```python
unit_cost = aggregate_frame['cost','sum'].div(aggregate_frame['speed','sum'],axis=0)
```

Since the DV is cost per unit bps (speed), it should be independent of absolute traffic volumes. More specifically, its reasonable to assume that unit cost would depend on the percentage of speed encountered at various levels. Convert all "speed" measures to ratio levels.

>traffic_ratio = speed (level)/total speed

```python
traffic_ratio = sparse_frame['speed','sum'].div(aggregate_frame['speed','sum'],axis=0)
```

Here is another measure of weighted average hops
> wtd_hops = hopspeed (level)/speed

```python
wtd_hops = aggregate_frame['hopspeed','sum'].div(aggregate_frame['speed','sum'],axis=0)
```

Start with plotting out a few variables.

```python
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(121)
ax1.hist(np.log(unit_cost),bins=20)
ax2 = fig.add_subplot(122)
ax2.hist(np.log(wtd_hops),bins=20)
```

The histogram on unit cost is evidence of a mixture, wherein three well seperated distributions exist. This hints at using a categorical predictor to seperate out those cases and fit individual models.

```python
pd.concat(dict(unit_cost = np.log(unit_cost),median_hops = aggregate_frame['hops','median']),axis=1).boxplot(by="median_hops",figsize = (10,5))
```

Lets create a dataset for performing regression on *unit_cost (dv)* with *wtd_hops (iv)* :

```python
from sklearn import preprocessing

data = pd.concat(dict(unit_cost=np.log(unit_cost),wtd_hops=np.log(wtd_hops-1)),axis=1).reset_index()
data[['unit_cost','wtd_hops']] = preprocessing.scale(data[['unit_cost','wtd_hops']])
data['country'] = data['level_0']
data['region'] = data['level_1']
data['pop'] = data['level_2']
data['rtr'] = data['level_3']
data['asn'] = data['level_4']
data.set_index(['level_0','level_1','level_2','level_3','level_4'],inplace=True)
```

Here is the distribution of number of customer-interfaces (Y) connected at a country/region/pop/router (X):

```python
fig = plt.figure(figsize = (17,12))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
data['country'].value_counts().plot(kind="bar",ax=ax1)
data['region'].value_counts().plot(kind="bar",ax=ax2)
data['pop'].value_counts().plot(kind="bar",ax=ax3)
data['rtr'].value_counts().plot(kind="bar",ax=ax4)
```

Clearly, some levels in country/region/pop (eg. CENTRAL AMERICA, AUSTRALIA) are relatively under sampled for any estimation procedure to fit reasonably well. One may either introduce shrinkage/penalization with random effects on the parameter estimates OR remove all such cases.

It turns out that using reference country as a categorical predictor discriminates the observations showing similar pattern with wtd_hops. Lets use this information to deploy a linear regression with *wtd_hops* (Continuous) and *country* (Categorical) predictors.

Note: reference country implies the entity which is considered from the cost liability perspective while sending/receiving traffic.

```python
import statsmodels.api as sm
import patsy
levels = ['NORTH AMERICA','EUROPE','ASIA','SOUTH AMERICA','CENTRAL AMERICA','AUSTRALIA']
formula = 'unit_cost~wtd_hops*C(country,levels = levels)'
y,X = patsy.dmatrices(formula,data,return_type='dataframe')
mod = sm.OLS(y,X).fit()
print mod.summary()
```

Here is a plot of the distinct regression lines:

```python
fig = plt.figure(figsize = (10,6))
groups = data.groupby(['country'])
country = pd.unique(data['country']).tolist()
colors = ['r','g','b','c','m','k']
for value,group in groups:
    idx = group.index
    plt.scatter(data['wtd_hops'][idx],data['unit_cost'][idx],c=colors[country.index(value)],s=100,alpha=0.3)
    plt.xlabel("wtd_hops")
    plt.ylabel("unit_cost")
    plot, = plt.plot(data['wtd_hops'][idx],mod.fittedvalues[idx],c=colors[country.index(value)],ls='-',label=value)
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

## Influential Observations

Influence plots provide an assessment of the influence of each observation. Usually, this is done by dropping an entire case from the dataset and refitting the model. If the case-omitted model is very different than the one using all the data then the particular case is influential for estimating the regression function. There are various standard measures of influence.

**DFFITS** measures how much the regression function changes at the i-th observation when the i-th observation is deleted. For small/medium datasets, a value of 1 or greater is considered "suspicious". For larger datasets, a value greater than $ 2\sqrt{(p+1)/n} $ is generally used.

```python
%%R -h 600 -w 800
plot(dffits(mod), pch=21, bg='blue', cex=1, ylab="DFFITS")
```

```python
%%R
print(Rdata[which(dffits(mod) > 1),])
```

**Cook’s distance** measures how much the entire regression function changes when the i-th case is deleted.

```python
%%R -h 600 -w 800
plot(cooks.distance(mod), pch=21, bg='blue', cex=1, ylab="Cook's distance")
print(Rdata[which(cooks.distance(mod) > 1),])
```

Similar cases are encountered as both DFFITS and Cook's distance measure changes in fitted values. The difference is that one measures the influence on one fitted value, while the other measures the influence on the entire vector of fitted values.

We may also exploit R's internal influencial point flagging rules: call influence.measures for a summary.

## Outliers

The essential definition of an outlier is an observation that does not follow the model, while most other observations seem to follow the model. A more rigorous definition that we are using is:

> A datum is considered to be an outlier if it will not ﬁt the “true” model instantiated by the “true” set of parameters within some error threshold that deﬁnes the maximum deviation attributable to the eﬀects of noise.

### Outlier in predictors:

The X values of the observation may lie outside the “cloud” of other X values. This means that one is extrapolating the model inappropriately. One way to detect outliers in the predictors, besides just looking at the actual values themselves, is through their leverage values, defined by the diagonal elements in the Hat Matrix.

```python
%%R -h 600 -w 800
plot(hatvalues(mod), pch=21, bg='blue', cex=1, ylab='Hat values')
print(Rdata[which(hatvalues(mod)>0.1),])
```

As evident in the scatterplots above, those levels of *country* where there are very few observations are particularly high leverage. This may cause problems in the parameter estimation for those levels, there are two ways to deal with this problem:

* Introduce penalization/random effects on the parameter estimates, which in our case boils down to effectively implementing a multilevel model
* Remove the cases where we encounter the situation of under represented levels apriori

### Outlier in response:

The Y value of the observation may lie very far from the fitted model, i.e. if the studentized residuals are large, the observation may be an outlier.

* Consider a crude outlier test that tries to find residuals that are larger than they should be. 
* Since studentized residuals are t distributed, we could just compare them to the T distribution and reject if their absolute value is too large. 

Doing this for every observation results in n different hypothesis tests, this causes a problem: if n is large, one will get many outliers by chance even if model is correct due to Type I errors. Every large data set would have outliers flagged, despite the model being entirely correct !

#### Multiple comparisons

This problem we identified is known as multiple comparisons or simultaneous inference. Perform n hypothesis tests while still controlling the probability of obtaining false positives.

The solution is to divide the p-value cutoffs by n, the number of tests. This is known as a Bonferroni correction, if we are doing many t (or other) tests, say m, we can control overall false positive rate at $\alpha$ by testing each one at level $\alpha/m$.

```python
%%R
n = nrow(Rdata)
cutoff = qt(1 - 0.001/ (2*n), (n - 4))
outliers = Rdata[which(abs(rstudent(mod)) > cutoff),]
print(outliers)
```

This plot represents residuals VS leverage, observations having large values in both dimensions are deemed influential outliers.

```python
%%R -h 600 -w 800
plot(mod, pch=21, bg='blue', cex=1, which=5)
```

Lets dig a little deeper into the cases found above:

```python
import statsmodels.stats.outliers_influence as infl
influence  = infl.OLSInfluence(mod)
diagnose = influence.summary_frame()
```

```python
%Rpull cutoff
data['dffits'] = 0
data['cooksd'] = 0
data['leverage'] = 0
data['rstudent'] = 0

data.loc[abs(diagnose.dffits)>1,"dffits"] = 1
data.loc[abs(diagnose.dffits)>1,"cooksd"] = 1
data.loc[abs(diagnose.hat_diag)>0.1,"leverage"] = 1
data.loc[abs(diagnose.student_resid)> cutoff[0],"rstudent"] = 1
```

Mark all outliers/influential cases as obtained from the statistics described above, below is a visualization of those points of interest (with big markers). A case-by-case further inspection should be worthwhile in understanding why such points have been flagged.

```python
fig = plt.figure(figsize = (16,12))
ax1 = fig.add_subplot(221)
ax1.set_title('DFFITS')
ax1.scatter(data.wtd_hops,data.unit_cost,c=pd.Categorical(data.country).labels,s=np.exp(3+2*data.dffits),alpha = 0.6)
ax2 = fig.add_subplot(222)
ax2.set_title('COOKSD')
ax2.scatter(data.wtd_hops,data.unit_cost,c=pd.Categorical(data.country).labels,s=np.exp(3+2*data.cooksd),alpha = 0.6)
ax3 = fig.add_subplot(223)
ax3.set_title('High Leverage')
ax3.scatter(data.wtd_hops,data.unit_cost,c=pd.Categorical(data.country).labels,s=np.exp(3+2*data.leverage),alpha = 0.6)
ax4 = fig.add_subplot(224)
ax4.set_title('High Residuals')
ax4.scatter(data.wtd_hops,data.unit_cost,c=pd.Categorical(data.country).labels,s=np.exp(3+2*data.rstudent),alpha = 0.6)
```

The focus is to study outliers from a model perspective, i.e. ones with high residuals, to account for its peculiar behaviour within the model context. Consider cases by reference country - ASIA, NORTH AMERICA, EUROPE and SOUTH AMERICA levels only for sufficient data.

```python
top_country = top_frame['speed','sum','country'][['value1','label1','value2','label2']]
top_rtr = top_frame['speed','sum','rtr'][['value1','label1','value2','label2']]
deep = data.join(top_country).join(top_rtr,rsuffix = "_rtr")
```

```python
fig = plt.figure(figsize = (16,12))
fig.suptitle("NORTH AMERICA")
check = deep.ix[('NORTH AMERICA')]
ax1 = fig.add_subplot(221)
ax1.scatter(check['wtd_hops'],check['unit_cost'],c=check.rstudent,s=60,alpha=0.6)
ax1.set_title('Scatter plot with outlier flagging')
ax2 = fig.add_subplot(222)
check['value1'].hist(ax=ax2)
ax2.set_title('Percentage Traffic to Top Country')
ax3 = fig.add_subplot(223)
check['label1'].value_counts().plot(kind="bar")
ax3.set_title('Top Country')
ax4 = fig.add_subplot(224)
check['value2'].hist(ax=ax4)
ax4.set_title('Percentage Traffic to Next Top Country')
check[check['rstudent'] == 1]
```

The above plots showcase that most customers connected in NORTH AMERICA have the tendency to interact with other customers within NORTH AMERICA, thus constituting local traffic. As represented by the histogram of top country traffic and the top country label, some 99.5% customers interact with other customers in NORTH AMERICA >95% of the time. 

The case (175,1239) is flagged on account of its interaction with customers in ASIA (approx. 7% by volume) as opposed to the cases in its neighborhood, all of which interact with local customers only.

```python
fig = plt.figure(figsize = (16,6))
fig.suptitle("EUROPE")
check = deep.ix[('EUROPE')]
ax1 = fig.add_subplot(121)
ax1.scatter(check['wtd_hops'],check['unit_cost'],c=check.rstudent,s=60,alpha=0.6)
ax1.set_title('Scatter plot with outlier flagging')
ax2 = fig.add_subplot(122)
check['value1'].hist(ax=ax2)
ax2.set_title('Percentage Traffic to Top Country')
fig, (ax3, ax4) = plt.subplots(figsize = (16,6),nrows=1, ncols=2)
check['label1'].value_counts().plot(kind="bar",ax=ax3).set_title("Top Country")
# hack
a = check[['wtd_hops','label1']]
b = pd.get_dummies(a['label1'])
# plot
b.replace(0,np.nan).mul(a['wtd_hops'],axis=0).boxplot(ax=ax4)
ax4.set_title("Wtd_hops by Other Country")
check[check['rstudent'] == 1]
```

For EU customers, we may conclude on similar lines that there is a general traffic pattern followed by >90% of customers. This general pattern shows that traffic to the top countries (NORTH AMERICA: 50% ; EUROPE: 50%) is mostly 100%, i.e. most customers interact with customers either only in EUROPE (~50%) OR only in NORTH AMERICA (~50%) as evident in the histograms and bar plots above. Also, the rest 10% of customers interact uniformly (50%-90%) with others in NORTH AMERICA/EUROPE. There exists a high degree of skewness in the countrywise customer interaction patterns.

Also evident from the Box Plot above, the wtd_hops to/from EUROPE customers will be small as compared to wtd_hops to NORTH AMERICA since the customers under consideration are all located in EUROPE.

Considering the outliers as indicated above, there is clear evidence that they deviate from this general pattern. (152,1239) shows a traffic mix of 57% to NORTH AMERICA and 33% to ASIA. This increases unit_cost as EUROPE-ASIA links are more expensive.

The other customer (154,6453) is a little hazy in its flagging, although it follows the general pattern (~97% to NORTH AMERICA), its wtd_hops (= -1.43) is quite small as opposed to the general median of (wtd_hops = 1.5) for interaction with customers in NORTH AMERICA. Hence, given the wtd_hops for (154,6453) are small, the fact that 97% of its traffic is actually to/from NORTH AMERICA renders it an outlier.
